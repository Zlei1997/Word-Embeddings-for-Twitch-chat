```python
import gensim
import re
import string
import pandas as pd
import tempfile
# some NLP libraries that can help with preprocessing
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df = pd.read_table('admiralbahroo.txt')
df
```

```python
df['newtext']  = df['body'].str.lower()
df
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

def stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["newtext"] = df["newtext"].apply(stopwords)
df
```

```python
punct = string.punctuation
punct = punct.replace("@", "") # don't remove @
punct = punct.replace("_","") # don't remove @
pattern = r"[{}]".format(punct) # create the pattern
```

```python
def remove_punct(x):
    return x.translate(str.maketrans("","",punct))
df['newtext']=df['newtext'].apply(lambda x: remove_punct(x))
df
```

```python
#Creating function for tokenization
def tokenization(text):
    text = re.split('\W+', text)
    return text
# Passing the function to 'text_lower' and store into'text_lower'
df['newtext'] = df['newtext'].apply(lambda x: tokenization(x.lower()))
df
```

```python
data = df['newtext']
data
```

```python
model = gensim.models.Word2Vec(
    window=100,
    min_count=0,
    workers=4
)
model.build_vocab(data, progress_per=1000)
model.train(data, total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar("hi")
```

```python
model.wv.most_similar(positive=['hi'], topn=20)
```

```python
with tempfile.NamedTemporaryFile(prefix='admiralbahroo-', delete=False) as tmp:
    temporary_filepath = tmp.name
    model.save(temporary_filepath)
```

```python
admiralbahroo = gensim.models.Word2Vec.load(temporary_filepath)
admiralbahroo.wv.most_similar(positive=['lul_emote'], topn=20)
```
